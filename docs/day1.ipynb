```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        print(" Starting execution...:",start_time)
        result = func(*args, **kwargs)
        
        end_time = time.time()
        print(" Ending execution...:",end_time)
        print(f"Execution time: {end_time - start_time} seconds")
        
        return result
    return wrapper


@timer
def add(a, b):
    return a + b


# Test
output = add(10, 20)
print("Result:", output)
```
