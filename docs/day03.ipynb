**Goal:** Build your first correct FastAPI service and understand request handling under the hood.

## 1) FastAPI request lifecycle

When a request reaches FastAPI:
1. Router matches path + method
2. Parameters parsed and validated
3. Dependencies resolved
4. Endpoint function executes
5. Response serialized to JSON

## 2) Minimal FastAPI app

```python
from fastapi import FastAPI

app = FastAPI(title="Learning API", version="1.0.0")

@app.get("/")
def root():
    return {"message": "API running"}
```

**Run:**
```bash
uvicorn main:app --reload
```

````python
# Simulating FastAPI request-response
class SimpleRoute:
    def __init__(self, path, method):
        self.path = path
        self.method = method
    
    def handle_request(self, incoming_path, incoming_method):
        if self.path == incoming_path and self.method == incoming_method:
            return {"matched": True, "path": self.path}
        return {"matched": False}

route = SimpleRoute("/users", "GET")
result = route.handle_request("/users", "GET")
print(f"Request result: {result}")
````

## 3) Why /docs is important

FastAPI auto-generates Swagger UI docs from your code.

**Benefits:**
- Quick API testing
- Shared contract with frontend
- Fewer documentation mismatches

Visit: `http://localhost:8000/docs`

## 4) Route patterns and order

**Static routes before dynamic routes!**

```python
@app.get("/users/me")          # static - comes first
def current_user():
    return {"user": "current"}

@app.get("/users/{user_id}")   # dynamic - comes second
def get_user(user_id: int):
    return {"id": user_id}
```

Why? Because `/users/me` would match `{user_id}` if dynamic route came first.

## 5) Response status codes

Be intentional:
- GET success: `200 OK`
- POST created: `201 Created`
- DELETE success (no body): `204 No Content`
- Missing resource: `404 Not Found`
- Invalid input: `422 Unprocessable Entity`

````python
from fastapi import FastAPI, HTTPException

# Simulating status code assignment
def handle_get_user(user_id):
    users = {1: {"id": 1, "name": "Alice"}}
    
    if user_id not in users:
        return {"status": 404, "detail": "User not found"}
    
    return {"status": 200, "data": users[user_id]}

print("User 1:", handle_get_user(1))
print("User 999:", handle_get_user(999))
````

## 6) Sync vs async in FastAPI

- Use `def` for sync work (CPU-light, sync DB)
- Use `async def` for async I/O (async DB, HTTP calls)

**Don't use async everywhere without reason.**
