# Day 2 - HTTP and REST Deep Dive

**Goal:** Understand exactly how APIs communicate using HTTP and design clean REST endpoints.

## 1) HTTP basics

HTTP is the protocol between client and server.

Every exchange has:
- **Request:** method + URL + headers + optional body
- **Response:** status code + headers + optional body

## 2) Request anatomy

Example:
```http
POST /users HTTP/1.1
Host: api.example.com
Content-Type: application/json
Authorization: Bearer <token>

{
  "name": "Safwan",
  "email": "safwan@example.com"
}
```

**Key parts:**
- Method: action (GET, POST, PUT, PATCH, DELETE)
- Path: resource target
- Headers: metadata
- Body: payload for create/update

## 3) Response anatomy

Example:
```http
HTTP/1.1 201 Created
Content-Type: application/json

{
  "id": 1,
  "name": "Safwan"
}
```

## 4) Status codes you must master

| Code | Meaning |
|------|----------|
| **200** | OK - request succeeded |
| **201** | Created - new resource made |
| **204** | No Content - success, no body |
| **400** | Bad Request - malformed |
| **401** | Unauthorized - missing auth |
| **403** | Forbidden - not allowed |
| **404** | Not Found - resource missing |
| **409** | Conflict - duplicate/state error |
| **422** | Unprocessable - validation failed |
| **500** | Internal Error - server bug |

## 5) REST design fundamentals

**Use nouns for resources:**
```
GET    /users        # list users
GET    /users/1      # get user 1
POST   /users        # create user
PATCH  /users/1      # update user 1
DELETE /users/1      # delete user 1
```

**Avoid action-style URLs:**
```
❌ /createUser
❌ /deleteUser
```

## 6) Idempotency and method semantics

| Method | Modifies state? | Idempotent? | Use |
|--------|-----------------|------------|------|
| GET | No | Yes | Read |
| PUT | Yes | Yes | Replace |
| PATCH | Yes | No | Partial update |
| DELETE | Yes | Yes (usually) | Delete |
| POST | Yes | No | Create |

**Idempotent:** Same call repeated = same effect

## 7) Query parameters for reads

Use query params for:
- **Filtering:** `/users?role=admin`
- **Sorting:** `/users?sort=created_at`
- **Pagination:** `/users?skip=0&limit=20`

````python
# Example: URLs with query parameters

urls = [
    "GET /users",
    "GET /users?skip=0&limit=10",
    "GET /users?role=admin&sort=name",
    "POST /users",
    "PATCH /users/5",
    "DELETE /users/5",
]

for url in urls:
    print(url)
````
