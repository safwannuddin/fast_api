## Goal
Write comprehensive tests that catch bugs before production. Master pytest, test databases, and mocking external services.

## Why This Day Matters
Code without tests breaks. Tests catch regressions, document behavior, and give confidence to refactor.

Measure code tested.

## PART 7: BEST PRACTICES

### ✅ DO:
1. Test happy path AND error cases
2. Use fixtures for reusable setup
3. Run full test suite before commit
4. Write tests for bug fixes (prevent regression)
5. Use fixtures instead of setUp/tearDown

### ❌ DON'T:
1. Test implementation details (too fragile)
2. Mock too much (defeats the purpose)
3. Leave test database with real data
4. Skip tests because they're slow (fix the speed!)
5. Write tests that pass sometimes (fix flakiness!)
