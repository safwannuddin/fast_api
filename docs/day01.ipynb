# Day 1 - Backend Foundations for Absolute Beginners

**Goal:** Understand what backend development is, how the web works end-to-end, and how FastAPI fits into real software systems.

## 1) What is backend development?

Backend is the part of software that handles data, business logic, security, and communication with databases or external services.

**If frontend is what users see, backend is the system that decides:**
- what data is valid
- who is allowed to do what
- where data is stored
- what response should be sent back

## 2) How a web request actually works

When a user clicks "Login" in an app:
1. Browser/mobile app sends an HTTP request
2. Backend receives request at an endpoint
3. Backend validates input
4. Backend checks DB or external services
5. Backend creates response (success/error)
6. Frontend renders response

**This request-response loop is the core of backend engineering.**

## 3) Key backend terms

| Term | Meaning |
|------|----------|
| **Client** | App that sends request |
| **Server** | App that receives & responds |
| **API** | Rules for client-server talk |
| **Endpoint** | Route like /users or /orders/5 |
| **Database** | Persistent data storage |
| **JSON** | Data format for APIs |
| **Stateless** | Server doesn't remember unless stored |

## 4) FastAPI in backend ecosystem

FastAPI is a Python framework to build APIs with type hints.

**Why teams use it:**
- ⚡ Fast performance
- ✅ Automatic validation
- 📖 Auto-generated docs
- 😊 Clean developer experience

## 5) Python concepts you need

Be comfortable with:
- functions
- dictionaries and lists
- classes
- type hints
- imports and virtual environments

```python
# Example: Function with type hints
def add(a: int, b: int) -> int:
    """Add two integers."""
    return a + b

result = add(5, 3)
print(f"5 + 3 = {result}")
```

Type hints are **powerful in FastAPI** because they drive validation and docs.

## 8) Practice exercises

1. Write 3 Python functions with type hints

3. Explain: What happens when you open a URL?

```python
# Exercise 1: Functions with type hints
def multiply(x: int, y: int) -> int:
    return x * y

def greet(name: str) -> str:
    return f"Hello, {name}!"

def is_adult(age: int) -> bool:
    return age >= 18

print(multiply(4, 5))
print(greet("Safwan"))
print(f"Is 25 an adult? {is_adult(25)}")
```

```python
# Exercise 2: Dictionary and list
user = {
    "id": 1,
    "name": "Safwan",
    "email": "safwan@example.com",
    "age": 25
}

users = [
    {"id": 1, "name": "Alice", "role": "admin"},
    {"id": 2, "name": "Bob", "role": "user"},
    {"id": 3, "name": "Charlie", "role": "user"},
]

print("Single user:", user)
print("\nAll users:")
for u in users:
    print(f"  - {u['name']} ({u['role']})")
```
